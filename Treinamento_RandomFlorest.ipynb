```python
arquivo = arquivo[arquivo['STATE'].notna()]
arquivo = arquivo[arquivo['CATEGORIA_TUR'].notna()]
arquivo = arquivo[arquivo['RURAL_URBAN'].notna()]
encode = pd.concat([pd.get_dummies(arquivo['STATE']), pd.get_dummies(arquivo['CATEGORIA_TUR']), pd.get_dummies(arquivo['RURAL_URBAN'])], axis=1)
arquivo.drop(columns=['STATE', 'CATEGORIA_TUR', 'RURAL_URBAN'], inplace=True)
arquivo = pd.concat([arquivo, encode], axis=1)
arquivo.head()
```

```python
arquivo['ALT'] = pd.to_numeric(arquivo['ALT'])
arquivo['IDHM'] = pd.to_numeric(arquivo['IDHM'])
arquivo['AREA'] = pd.to_numeric(arquivo['AREA'].str.replace(',', ''))
arquivo['GVA_AGROPEC'] = pd.to_numeric(arquivo['GVA_AGROPEC'])
arquivo['GVA_INDUSTRY'] = pd.to_numeric(arquivo['GVA_INDUSTRY'])
arquivo['GVA_SERVICES'] = pd.to_numeric(arquivo['GVA_SERVICES'])
arquivo['GVA_PUBLIC'] = pd.to_numeric(arquivo['GVA_PUBLIC'])
arquivo[' GVA_TOTAL '] = pd.to_numeric(arquivo[' GVA_TOTAL '])
arquivo['TAXES'] = pd.to_numeric(arquivo['TAXES'])
arquivo['GDP'] = pd.to_numeric(arquivo['GDP'])
arquivo['GDP_CAPITA'] = pd.to_numeric(arquivo['GDP_CAPITA'])
```

```python
y = arquivo['IDHM']
x = arquivo.drop(columns='IDHM')
```

```python
from sklearn.preprocessing import MinMaxScaler
normalizador = MinMaxScaler(feature_range=(0, 1))
x_norm = normalizador.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
x_pca = pca.fit_transform(x_norm)
print(pca.explained_variance_ratio_.sum())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

florest_parans = {'max_depth': range(10, 13), 'min_samples_split': range(2, 5), 'min_samples_leaf': range(1, 4)}

modelo = RandomForestRegressor(n_estimators=200, n_jobs=-1)

gridFlorest = GridSearchCV(estimator=modelo, param_grid=florest_parans, cv=5, n_jobs=-1)

gridFlorest.fit(x_pca, y)

print('Melhor Acurácia: ', gridFlorest.best_score_) 
print('Melhor max_depth: ', gridFlorest.best_estimator_.max_depth)
print('Melhor min_samples_split: ', gridFlorest.best_estimator_.min_samples_split)
print('Melhor min_samples_leaf: ', gridFlorest.best_estimator_.min_samples_leaf)
```
